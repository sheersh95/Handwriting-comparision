```python
import pandas as pd
import sklearn.utils as sk
```

```python
with open("GSC-Features.csv", "r") as f:
    x = pd.read_csv(f)
```

```python
with open("diffn_pairs_gsc.csv", "r") as g:
    y = pd.read_csv(g)
```

```python
with open("same_pairs_gsc.csv", "r") as g:
    z = pd.read_csv(g)
```

```python
y = sk.shuffle(y).reset_index(drop=True)
z = sk.shuffle(z).reset_index(drop=True)
y1 = y[0:2000]
z1 = z[0:2000]
Raw = pd.concat([y1,z1],ignore_index = True)
df = pd.DataFrame(columns = range(0,1025))
```

```python
for i in range(Raw['img_id_A'].shape[0]):
    f1 = Raw['img_id_A'][i]
    f2 = Raw['img_id_B'][i]
    foc1 = x.loc[x['img_id'] == f1]
    foc2 = x.loc[x['img_id'] == f2]
    target = Raw['target'][i]
    target = pd.Series(target)
    foc1 = foc1.iloc[0,1:513]
    foc2 = foc2.iloc[0,1:513]
    com = pd.concat([foc1,foc2,target],ignore_index = True)
    com = com.reshape(1,1025)
    df.loc[i] = com[0]
```

```python
df.to_csv("GSCFeatureData.csv")
```
